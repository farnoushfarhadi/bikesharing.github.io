### Prediction of Bike Sharing Demand using RandomForest Method

The goal of this project is to predict the bicycle demands for
both casual and registered customers so that the bike sharing
system providers are be able to setup the system and also the
users plan their travels more easily.

### Raw Data: 

The data set provided by Kaggle demonstrates hourly rental
data over two years (from January 1, 2011 to December 31,
2012). The training data set is for the first 19 days of each month.
The remaining days are considered as test data. We aim to
predict the total rental bikes counts hourly in test data. The
date and time are provided as a string feature in the data

### Data Preprocessing:

To understand the distribution of variables, we plot a

After deeping dive in the data set, we have the hypothesis

As understood from following histograms, the season feature
has four categories of almost same distribution, and weather
attribute with partly cloudy or clear state is more frequent
than other weather conditions.
Furthermore, a considerable number of demands are received

weekend nor holiday.

For feature engineering, we perform the code below:


A logarithmic transformation of dependent cols is performed by adding 1 first so that 0 values don't become -inf.
Above function parses datetime colum and add new time related columns.


```python
def feature_engineering(df):
    # parse datetime colum & add new time related columns
    dt = pd.DatetimeIndex(df['datetime'])
    df['date'] = dt.date
    df['day'] = dt.day
    df['month'] = dt.month
    df['year'] = dt.year
    df['hour'] = dt.hour
    df['dow'] = dt.dayofweek
    df['woy'] = dt.weekofyear
    df['weeks_since_start'] = (dt - min(dt)).days / 7

    for col in ['casual', 'registered', 'count']:
        df['%s_log' % col] = np.log(df[col] + 1)
    return df
```

```python
def Data_Analysis_HistPlot():
    pd.set_option('display.width', 500)
    pd.set_option('display.max_columns', 30)

    dfs = {}
    for name in ['train', 'test']:
        df = pd.read_csv('D:/MASTER-POLY/trimister 3-Winter 2016/IND6212-Exploration des Donnees/project/codes/data/%s.csv' % name)
        df['_data'] = name
        dfs[name] = df

    # combine train and test data into one df
    df = dfs['train'].append(dfs['test'])

    # lowercase column names
    df.columns = map(str.lower, df.columns)
    df.head()
    df = feature_engineering(df)

    df.iloc[0]
    
    # get training data
    dg = df[df['_data'] == 'train'].copy()

    # distribution of casual / registered / count rentals by day
    by_day = dg.groupby('dow')[['casual', 'registered', 'count']].agg('sum')
    print by_day.describe()
    by_day.plot(kind='bar', figsize=(8,4), width=0.8, title = 'Distribution of Bike Rentals by Day')
    
    # distribution of casual / registered / count rentals by hour
    by_hour = dg.groupby('hour')[['casual', 'registered', 'count']].agg('sum')
    print by_hour.describe()
    by_hour.plot(kind='bar', figsize=(8,4), width=0.8, title = 'Distribution of Bike Rentals by Hour')

    # distribution of casual / registered / count rentals by season
    by_season = dg.groupby('season')[['casual', 'registered', 'count']].agg('sum')
    print by_season.describe()
    by_season.plot(kind='bar', figsize=(8,4), width=0.8, title = 'Distribution of Bike Rentals by Season')

    # distribution of casual / registered / count rentals by weather
    by_weather = dg.groupby('weather')[['casual', 'registered', 'count']].agg('sum')
    print by_weather.describe()
    by_weather.plot(kind='bar', figsize=(8,4), width=0.8, title = 'Distribution of Bike Rentals by Weather')

    # distribution of casual / registered / count rentals by humidity
    by_humidity = dg.groupby('humidity')[['casual', 'registered', 'count']].agg('sum')
    print by_humidity.describe()
    by_humidity.plot(kind='bar', figsize=(8,4), width=0.8, title = 'Distribution of Bike Rentals by Humidity')

    # distribution of casual / registered / count rentals by holiday
    by_holiday = dg.groupby('holiday')[['casual', 'registered', 'count']].agg('sum')
    print by_holiday.describe()
    by_holiday.plot(kind='bar', figsize=(8,4), width=0.8, title = 'Distribution of Bike Rentals by Holiday')

    # distribution of casual / registered / count rentals by temp
    by_temp = dg.groupby('temp')[['casual', 'registered', 'count']].agg('sum')
    print by_temp.describe()
    by_temp.plot(kind='bar', figsize=(8,4), width=0.8, title = 'Distribution of Bike Rentals by Temperature')

    # distribution of casual / registered / count rentals by atemp
    by_atemp = dg.groupby('atemp')[['casual', 'registered', 'count']].agg('sum')
    print by_atemp.describe()
    by_atemp.plot(kind='bar', figsize=(8,4), width=0.8, title = 'Distribution of Bike Rentals by aTemperature')

    # distribution of casual / registered / count rentals by workingday
    by_workingday = dg.groupby('workingday')[['casual', 'registered', 'count']].agg('sum')
    print by_workingday.describe()
    by_workingday.plot(kind='bar', figsize=(8,4), width=0.8, title = 'Distribution of Bike Rentals by Workingday')

    # distribution of casual / registered / count rentals by windspeed
    by_windspeed = dg.groupby('windspeed')[['casual', 'registered', 'count']].agg('sum')
    print by_windspeed.describe()
    by_windspeed.plot(kind='bar', figsize=(8,4), width=0.8, title = 'Distribution of Bike Rentals by Wind Speed')
          
    
    # total rentals by month & year
    z = dg.groupby(['year', 'month'])['count'].agg('sum').unstack()
    z.plot(kind='bar', figsize=(8,4), width=0.8)

    ############# correlation matrix #############
    cols = ['count', 'casual', 'registered', 'weather', 'temp', 'atemp', 
            'humidity', 'windspeed', 'holiday', 'workingday', 'season', 
            'day', 'month', 'year', 'hour', 'dow', 'woy']

    cm = dg[cols].corr()
    # plotting the correlation matrix
    cm = cm.astype('float')
    plt.matshow(cm)
    plt.title('Confusion matrix')
    plt.colorbar()
    x1 = range(0, len(cols))
    plt.xticks(x1, cols)
    plt.yticks(x1, cols)
    plt.show()
    return dg
```
